```python
import numpy as np 
import random
import networkx as nx
import matplotlib.pyplot as plt
import time 
```

```python
test_fix=[]
for i in range (1000):
    test_fix.append(np.loadtxt('fixtime_r3_'+str(i)+'.dat'))
```

```python
f3_mean=[]
for f in test_fix:
    f3_mean.append(np.sum(f)/len(f))
```

```python
plt.plot(f3_mean,'orange')
plt.xlabel('number of added edges to graph');plt.ylabel('fixation time_r3')
plt.ylim(0,350000)
```

```python
np.savetxt('f2.dat',f3_mean)
```

```python
from numpy import arange
from scipy.optimize import curve_fit
plt.loglog(f3_mean,'orange')
popt,pcov=curve_fit(lambda x,c,a:c*x**(a),np.arange(0,1000)[5:],f3_mean[5:])
plt.plot(np.arange(0,1000)[5:],popt[0]*np.arange(0,1000)[5:]**(popt[1]))
plt.ylim(0,1000000)
plt.xlim(0,10000)
plt.xlabel('Log number of added edges to graph');plt.ylabel('Log fixation time r=3')
plt.legend()
plt.show()
```
